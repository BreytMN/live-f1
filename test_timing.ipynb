```python
def build_timing_table(df):

    timing_df = df.copy()

    timing_df["InPit"] = timing_df["InPit"].fillna(False).astype(int).cumsum()
    return timing_df[["timestamp","NumberOfLaps","NumberOfPitStops","InPit"]]

build_timing_table(df_11)
```
