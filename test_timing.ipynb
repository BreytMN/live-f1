```python
cols = ["LastLapTime_Value","Sectors_0_Value","Sectors_1_Value","Sectors_2_Value","Sectors_0_PreviousValue","Sectors_1_PreviousValue","Sectors_2_PreviousValue","NumberOfLaps","InPit","PitOut"]

def str_timedelta(x):
    if isinstance(x,str):
        count_sep = x.count(":")

        if count_sep == 0:
            return "00:00:" + x
        elif count_sep == 1:
            return "00:" + x
        else: return x
    else:
        return x


df_test = df_11[["timestamp"] + cols ].dropna(subset=cols, how="all").replace('',np.nan)

# df_test["LapElapsed"] = df_test["Sectors_0_Value"] + df_test["Sectors_1_Value"] + df_test["Sectors_2_Value"]
# for col in df_test.columns:
#     df_test[col] = pd.to_timedelta(df_test[col])


for col in ["Sectors_0_Value","Sectors_1_Value","Sectors_2_Value", "Sectors_0_PreviousValue","Sectors_1_PreviousValue","Sectors_2_PreviousValue", "LastLapTime_Value"]:
    df_test[col] = df_test[col]
    # df_test[col] = df_test[col].fillna("00:00:00.000")
    df_test[col] = pd.to_timedelta(df_test[col].apply(str_timedelta))

# for col in []:
#     df_test[col] = df_test[col]
#     # df_test[col] = df_test[col].fillna("00:00:00.000")
#     df_test[col] = pd.to_timedelta(df_test[col].apply(str_timedelta))


df_test["LapElapsed"] = df_test["Sectors_0_Value"] + df_test["Sectors_1_Value"] + df_test["Sectors_2_Value"]


df_test.iloc[50:100]
# df_test.head(50)
```

```python
sector_cols = [
    "Sectors_0_Value",
    "Sectors_1_Value",
    "Sectors_2_Value",
    "Sectors_0_PreviousValue",
    "Sectors_1_PreviousValue",
    "Sectors_2_PreviousValue"
]

cols = ["lap_no","1","2","3", "LapTime"]

def enter_new_lap(laps, record):
    print(">>", record)
    if laps is None and record is None:
        return [], {key:None if key != "lap_no" else 1 for key in cols}, timedelta(seconds=0)

    if record["LapTime"] is None:
        record["LapTime"] = record["1"] + record["2"] + record["3"]

    laps.append(record)
    record = {key:None if key != "lap_no" else val+1 for key, val in record.items()}
    return laps, record


new_lap_allowed = True
laps, record, last_record_ts = enter_new_lap(None, None)

for idx,row in df_test.iterrows():
    ts = pd.to_timedelta(row.timestamp)
    
    for sc_key, sc_value in row[["LastLapTime_Value"]].dropna().to_dict().items():
        record["LapTime"] = sc_value
        
    for sc_key, sc_value in row[sector_cols].dropna().to_dict().items():
        sc_no = int(sc_key.split("_")[1])
        key_type = sc_key.split("_")[2]
        print(sc_no, "||", ts, "||", key_type, sc_value)

        if key_type == "Value":
            if record[str(sc_no + 1)] == None:
                record[str(sc_no + 1)] = sc_value
                last_record_ts = ts
                if sc_no == 2:
                    # entering new lap
                    laps, record = enter_new_lap(laps, record)
            elif sc_value == record[str(sc_no + 1)]: pass
            elif ts - last_record_ts > timedelta(seconds=10):
                # entering new lap
                laps, record = enter_new_lap(laps, record)
                record[str(sc_no + 1)] = sc_value
                last_record_ts = ts

        elif key_type == "PreviousValue" and ts - last_record_ts > timedelta(seconds=10):
            record[str(sc_no + 1)] = sc_value
            last_record_ts = ts
            if sc_no == 2:
                # entering new lap
                laps, record = enter_new_lap(laps, record)

laps_df = pd.DataFrame(laps)

segments = ["1","2","3"]
for idx in range(len(segments)):
    rest = np.delete(segments, idx)
    laps_df[segments[idx]] = (laps_df[segments[idx]].fillna(timedelta(minutes=0)) + (laps_df[segments[idx]].isnull() & (laps_df["lap_no"] > 1)) * (laps_df[segments[idx]].isnull() * (laps_df["LapTime"].fillna(timedelta(minutes=0)) - laps_df[rest].sum(axis=1)))).replace(timedelta(minutes=0), np.timedelta64("NaT"))

laps_df
```
